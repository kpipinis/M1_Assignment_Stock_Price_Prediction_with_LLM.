## Step 3.1: Model Building - Linear Regression

```python
!pip install ta
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import ta
import pandas as pd # Ensure pandas is imported

# Assuming 'data' DataFrame is populated as described in the kernel state.

# Flatten multi-level columns from yfinance output for easier access
# The kernel state shows columns like ('Price', 'AAPL'), ('Close', 'AAPL'), etc.
if isinstance(data.columns, pd.MultiIndex):
    # Take the first level of the MultiIndex (e.g., 'Close', 'Price')
    data.columns = data.columns.get_level_values(0)

# 1. Feature Engineering

# Historical Prices (Open, High, Low, Close) and Volume are now directly accessible
# 'Close' often represents the adjusted close in yfinance data, especially with auto_adjust=True.

# Moving Averages
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()

# Technical Indicators: RSI and MACD
data['RSI'] = ta.momentum.RSIIndicator(data['Close'], window=14).rsi()
macd = ta.trend.MACD(data['Close'])
data['MACD_line'] = macd.macd()
data['MACD_signal'] = macd.macd_signal()

# Create the target variable: 'Tomorrow_Close'
# This is the 'Close' price of the next trading day
data['Tomorrow_Close'] = data['Close'].shift(-1)

# Drop rows with NaN values introduced by feature engineering and target variable creation
df_model = data.dropna()

# Define features (X) and target (y)
# Removed 'Price' from features as 'Close' is typically the adjusted price
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_5', 'SMA_10', 'SMA_20', 'SMA_50', 'RSI', 'MACD_line', 'MACD_signal']
X = df_model[features]
y = df_model['Tomorrow_Close']

# 2. Train-Test Split (70/30 chronological split)
split_index = int(len(df_model) * 0.7)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"Training set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")

# 3. Model Training (Linear Regression)
model = LinearRegression()
model.fit(X_train, y_train)

# 4. Model Evaluation
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

## Step 3.2: Model Building - Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Create the binary target variable: 1 if 'Close' increases tomorrow, 0 otherwise
data['Tomorrow_Direction'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Drop rows with NaN values introduced by feature engineering and new target variable
df_classifier_model = data.dropna()

# Define features (X) and the new target (y_direction)
# 'Tomorrow_Close' is no longer needed as a feature, but used to create 'Tomorrow_Direction'
features_clf = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_5', 'SMA_10', 'SMA_20', 'SMA_50', 'RSI', 'MACD_line', 'MACD_signal']
X_clf = df_classifier_model[features_clf]
y_direction = df_classifier_model['Tomorrow_Direction']

# 2. Train-Test Split (70/30 chronological split)
split_index_clf = int(len(df_classifier_model) * 0.7)

X_train_clf, X_test_clf = X_clf[:split_index_clf], X_clf[split_index_clf:]
y_train_direction, y_test_direction = y_direction[:split_index_clf], y_direction[split_index_clf:]

print(f"Training set size (Classifier): {len(X_train_clf)} samples")
print(f"Testing set size (Classifier): {len(X_test_clf)} samples")

# 3. Model Training (Random Forest Classifier)
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train_clf, y_train_direction)

# 4. Model Evaluation
y_pred_direction = clf_model.predict(X_test_clf)

accuracy = accuracy_score(y_test_direction, y_pred_direction)
conf_matrix = confusion_matrix(y_test_direction, y_pred_direction)
class_report = classification_report(y_test_direction, y_pred_direction)

print(f"\nAccuracy: {accuracy:.2f}")
print(f"\nConfusion Matrix:\n{conf_matrix}")
print(f"\nClassification Report:\n{class_report}")

# Feature Importance
feature_importances = pd.Series(clf_model.feature_importances_, index=features_clf).sort_values(ascending=False)
print(f"\nFeature Importance:\n{feature_importances}")
```
